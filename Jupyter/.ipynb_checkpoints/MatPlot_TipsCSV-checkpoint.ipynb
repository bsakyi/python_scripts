```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = pd.read_csv("tips.csv")
```

```python
tips.describe()
```

```python
tips["total_bill"].plot(kind="hist", )
```

```python
plt.hist(tips["total_bill"], bins=5, color='#b6cdf1', edgecolor='#144970')
plt.title('Distribution of Total Bill')   
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
```

```python
plt.scatter(tips['total_bill'], tips['tip'], color='coral')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
tips.groupby('day')['tip'].mean().plot(kind='bar', color='coral')
plt.title('Average Tip Amount by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip')
```

```python
# Heat Maps=> correlation between variables
correlation_matirx = tips[['total_bill', 'tip', 'size']].corr()
plt.imshow(correlation_matirx, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title('Correlation Matrix')
plt.show()
```
